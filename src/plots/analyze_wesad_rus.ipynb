```python
import sys
import numpy as np
import matplotlib.pyplot as plt
```

```python
# mimiciv_rus_results[0:6]
def plot_multi_lag_results(feature_pair_data, feature_pair_name, ax=None):
    """
    Plot RUS decomposition results for different lags of a feature pair.
    
    Args:
        feature_pair_data: Array of dictionaries containing RUS results for different lags
        feature_pair_name: String name of the feature pair for the title
        ax: Matplotlib axis object (optional)
    """
    if ax is None:
        fig, ax = plt.subplots(figsize=(10, 6))
    
    # Extract data for plotting
    lag_results = feature_pair_data['lag_results']
    lags = [item['lag'] for item in lag_results]
    r_values = [item['R_value'] for item in lag_results]
    u1_values = [item['U1_value'] for item in lag_results]
    u2_values = [item['U2_value'] for item in lag_results]
    s_values = [item['S_value'] for item in lag_results]
    mi_values = [item['MI_value'] for item in lag_results]
    
    # Create the line plot
    ax.plot(lags, r_values, '-', label='R (Redundancy)', linewidth=2)
    ax.plot(lags, u1_values, '-', label=f'U₁ (Unique {feature_pair_data["feature_pair"][0]})', linewidth=2)
    ax.plot(lags, u2_values, '-', label=f'U₂ (Unique {feature_pair_data["feature_pair"][1]})', linewidth=2)
    ax.plot(lags, s_values, '-', label='S (Synergy)', linewidth=2)
    ax.plot(lags, mi_values, '-', label='MI (Total)', linewidth=2, color='black', linestyle='--')
    
    # # Annotate the dominant term
    # for i, item in enumerate(feature_pair_data):
    #     dom_term = item['dominant_term']
    #     ax.annotate(dom_term, (lags[i], mi_values[i]), 
    #                xytext=(0, 5), textcoords='offset points',
    #                ha='center', fontweight='bold')
    
    # Add labels and title
    ax.set_xlabel('Lag (time steps)', fontsize=12)
    ax.set_ylabel('Information (bits)', fontsize=12)
    ax.set_title(f"RUS Decomposition\n {feature_pair_name}", fontsize=14)
    
    # Add legend and grid
    ax.legend(loc='best')
    ax.grid(True, linestyle='--', alpha=0.7)
    
    # Set y-axis to start at 0
    ax.set_ylim(bottom=0)
    
    return ax


```

```python
wrist_rus_path = '/home/hhchung/dami/results/wesad/rus_fine_grained_wrist_modality_lag100_bins8.npy'
```

```python
wrist_rus_results = np.load(wrist_rus_path, allow_pickle=True)
len(wrist_rus_results)
```

```python
# Extract the first feature pair and plot
pair_data = wrist_rus_results[0]
pair_name = f"X1: {pair_data['feature_pair'][0]} X2: {pair_data['feature_pair'][1]} Y: wrist_temp"
fig, ax = plt.subplots(figsize=(10, 6))
plot_multi_lag_results(pair_data, pair_name, ax)
plt.tight_layout()
plt.show()
```
